```python
categoricals = customer_df.select_dtypes(include="object")
categoricals.head(5)
```

```python
# We do not consider "Customer_ID" as theyÂ´re all different
# We drop "Customer"

new_categoricals = categoricals.drop(["Customer"],axis=1)

new_categoricals
```

```python
# Create hostogram
new_categoricals["State"].value_counts().plot(kind="bar")
```

```python
# Now we run in it for all categorical columns

cols = new_categoricals.columns

for i in cols:
    new_categoricals[i].value_counts().plot(kind="bar")
    plt.show()
    
```
