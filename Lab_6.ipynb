```python
# Lets choose a couple of variables: "income" & "monthly premium auto"

# First we create a function for log transformation

def log_transform_clean2(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN
```

```python
numericals['income_new'] = numericals['income'].apply(log_transform_clean2) 
numericals['monthly premium auto_new'] = numericals['monthly premium auto'].apply(log_transform_clean2) 
```

```python
# We Replace Nan with means
numericals['income_new'].isna().sum()
numericals['monthly premium auto_new'].isna().sum()
```

```python
# We replace Nan with means
numericals['income_new'] = numericals['income_new'].fillna(np.mean(numericals['income_new']))
numericals['monthly premium auto_new'] = numericals['monthly premium auto_new'].fillna(np.mean(numericals['monthly premium auto_new']))
```

```python
# Check
sns.displot(numericals['income_new'])
sns.displot(numericals['monthly premium auto_new'])
```
