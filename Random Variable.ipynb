```python
from scipy.stats import binom, poisson, uniform, norm, expon
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
```

```python
# Load data
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
numerical = customer_df.select_dtypes(include='number')
```

```python
numerical.columns
```

#### 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
   

```python
# Lets look at the Histogram plots
for i in numerical.columns:
    sns.displot(x = numerical[i])
    plt.show()
```

```python
# Lets look now to QQ-Plots
```

```python
for i in numerical.columns:
    sm.qqplot(numerical[i],line='s')
    plt.show()
```

```python
# No feature seems to have a Normal distribution as ditribution and QQ plots show
```

 #### 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
# Logarithmic Transformation


for i in numerical.columns:
    numerical[i] = np.log(numerical[i])
    sns.displot(x = numerical[i])
    sm.qqplot(numerical[i],line='s')
    plt.show()
```

```python
# We can see that some variables such a Customer Lifetime Value are now more "normal". Some others are still not normal
```

```python
# Square root Transformation


for i in numerical.columns:
    numerical[i] = np.sqrt(numerical[i])
    sns.displot(x = numerical[i])
    sm.qqplot(numerical[i],line='s')
    plt.show()
```

```python
# Similar results as with Log transformation
```

```python
# Lets try with Box-Cox


for i in numerical.columns:
    numerical[i],parameters=stats.boxcox(numerical[i])
    sns.displot(x = numerical[i])
    sm.qqplot(numerical[i],line='s')
    plt.show()
```

```python
# Even Box Cox is unable to normalize some of the features. On the other hand Customer Lifetime Value seems quite normal now.
```
